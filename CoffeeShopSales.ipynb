```python
import pandas as pd  #data manipulation
import numpy as np #calculations
import matplotlib.pyplot as plt #basic visualization
import seaborn as sns #adv visualization
```

```python
Sales = pd.read_excel("/Coffee Shop Sales.xlsx")
Sales
```

```python
Sales.info()
```

```python
Sales.describe()
```

```python
Sales.duplicated().sum()
```

```python
Sales.drop_duplicates()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='product_category', y='store_id', data=Sales, estimator=sum)
plt.xticks(rotation=45)
plt.title('Total Sales per Product')
plt.show()
```

```python
numeric_data = Sales.select_dtypes(include=['number'])
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(x='transaction_date', y='product_type', data=Sales)
plt.xticks(rotation=45)
plt.title('Sales Trend Over Time')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='store_location', y='unit_price', data=Sales, estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title('Total Sales per Region')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='transaction_qty', y='product_type', data=Sales)
plt.xticks(rotation=45)
plt.title('Sales Distribution by Category')
plt.show()
```
